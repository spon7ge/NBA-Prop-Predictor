```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
import undetected_chromedriver as uc
import pandas as pd
from datetime import datetime
```

```python
url = 'https://www.espn.com/nba/injuries'

print("Opening Browser...")
driver = webdriver.Firefox()
driver.get(url)

print("Waiting for page to load...")
try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "ResponsiveTable"))
    )
    print("Table found!")
except Exception as e:
    print(f"Table not found...")
    driver.quit()
    exit()

#contains all the injuries
tables = driver.find_elements(By.CLASS_NAME, "ResponsiveTable")

if not tables:
    print("No injuries found...")
    driver.quit()
    exit()

playerInjuries = []

for table in tables:
    rows = table.find_elements(By.TAG_NAME, "tr")[1:]
    for row in rows:
        cells = row.find_elements(By.TAG_NAME, "td")
        if len(cells) >= 5:
            player = cells[0].text.strip()
            position = cells[1].text.strip()
            est_return_date = cells[2].text.strip()
            status = cells[3].text.strip()
            comment = cells[4].text.strip()
            
            print(f"Extracted:{player} {position} {est_return_date} {status} {comment}")
            playerInjuries.append(
                [player, position, est_return_date, status, comment]
            )

injuriesDF = pd.DataFrame(playerInjuries, columns=['PLAYER', 'POSITION', 'EST. RETURN DATE', 'STATUS', 'COMMENT'])







```
